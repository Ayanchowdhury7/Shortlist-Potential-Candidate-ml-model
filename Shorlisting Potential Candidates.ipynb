```python
degree = []

for words in df['Degree'].values:
    
    if "m.tech" in words.lower():
        degree.append("M.tech")
    elif "m.sc" in words.lower():
        degree.append("M.Sc")
    elif "b.tech" in words.lower():
        degree.append("B.Tech")
    elif "pg" in words.lower():
        degree.append("PG")
    elif "b.e" in words.lower():
        degree.append("B.Tech")
    else:
        degree.append(np.nan)
        
```

```python
df.drop(columns = 'Degree', inplace = True)
Degree = pd.DataFrame(degree)
df.insert(7, "Degree", Degree)
```

```python
df.rename(columns = {"Python (out of 3)": "Python",
                     "R Programming (out of 3)" : "R",
                     "Data Science (out of 3)" : "DS",
                     "Current Year Of Graduation" : "Year"},
                          inplace = True)
```

```python
# Taking rows where Degree value is not na
df = df[df['Degree'].notna()]

# Taking rows where Year value is max 2020.
df = df.loc[df['Year'] <= 2020]
```

### Insights
After applying the given condition that, Applicant should have a M.Sc / M.Tech / B.Tech/ B.E / PG degree , and graduation year should be max year 2020. We can see that out of 611 Applications only 383 Applications are shorlisted. So 62% selected for the further analysis based on their Degree and Year of Graduation.

```python
def f(row):
    if row['Degree'] == 'B.Tech' and row['Year'] == 2020:
        val = 10
    elif row['Degree'] == 'B.Tech' and row['Year'] == 2019:
        val = 8
    elif row['Degree'] == 'B.Tech' and row['Year'] <= 2018:
        val = 5
    elif row['Degree'] == 'M.Tech' and row['Year'] == 2020:
        val = 7
    elif row['Degree'] == 'M.Tech' and row['Year'] <= 2019:
        val = 3
    elif row['Degree'] == 'M.Sc' and row['Year'] == 2020:
        val = 7
    elif row['Degree'] == 'M.Sc' and row['Year'] <= 2019:
        val = 3
    elif row['Degree'] == 'PG' and row['Year'] == 2020:
        val = 7
    elif row['Degree'] == 'PG' and row['Year'] <= 2019:
        val = 3
    else:
        val = 0
    return val
```

```python
df['Degree_Year_Weightage'] = df.apply(f, axis = 1)
```

# Data Modification

As per the conditions:
__________________________________________________

If Python Value is 1 then it's weightage value is 3  
If Python Value is 2 then it's weightage value is 7  
If Python Value is 3 then it's weightage value is 10  
__________________________________________________

If R Value is 1 then it's weightage value is 3  
If R Value is 2 then it's weightage value is 7  
If R Value is 3 then it's weightage value is 10  
__________________________________________________

If DS Value is 1 then it's weightage value is 3  
If DS Value is 2 then it's weightage value is 7  
If DS Value is 3 then it's weightage value is 10  
__________________________________________________

Also Other Skills (i.e - Machine Learning, Deep Learning, NLP, Statistical Data Analysis, AWS, SQL, NoSQL, Excel) holds 3 weightage for each.

```python
df['Python'].replace(3, 10, inplace = True)
df['Python'].replace(1, 3, inplace = True)
df['Python'].replace(2, 7, inplace = True)

df['R'].replace(3, 10, inplace = True)
df['R'].replace(1, 3, inplace = True)
df['R'].replace(2, 7, inplace = True)

df['DS'].replace(3, 10, inplace = True)
df['DS'].replace(1, 3, inplace = True)
df['DS'].replace(2, 7, inplace = True)
```

```python
df['skills'].replace(np.nan, 0.0, inplace = True)
df = df.astype({"skills" : int})
df['skills'] = df['skills'].apply(lambda x: x * 3)
```

```python
df.set_index(['Application_ID','Degree','Year'], inplace = True)
```

```python
# creating a new column 'Total Score' and sum the scores of Python, R, Data Science(DS),Skills weightage,Degree_year_weightage.

df['Total Score'] = df.sum(axis = 1)
```

# Final Condition

If the Total Score is >= 40 then the Applicants profile has been shortlisted for Data Scientist, else Applicants profile did not qualify for further discussion.

```python
def score(row):
    if row['Total Score'] >= 40:
        val = "Congratulations"
    else:
        val = "Sorry"
    return val
```

```python
df['Results'] = df.apply(score, axis = 1)
```
